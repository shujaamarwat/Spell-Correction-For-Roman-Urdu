Insert, Delete, Substitution, Transposition Table 

```python
def idst_table(word):
  alphabets = 'abcdefghijklmnopqrstuvwxyz'
  # "","karo"
  # "k","aro"
  # "ka","ro"
  # "kar","o"

  splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
  # print(splits)

  # "","karo" = "" + "aro" = aro
  # "k","aro" = "k" + "ro" = kro
  # "ka","ro" = "ka" + "o" = kao
  # "kar","o" = "kar" + "" = kar
  deletes    = [L + R[1:]               for L, R in splits if R]
  # print(deletes)

  # "","karo" = "" + a  + k + "ro" = akro
  # "k","aro" = "k" + r  + a + "o" = krao
  # "ka","ro" = "ka" + o  + r + "" = krao
  transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
  # print(transposes)

  # "","karo" = "" + b + "aro" = baro 
  # "k","aro" = "k" + b + "ro" = kbro
  # "ka","ro" = "ka" + b + "o" = kabo
  # "kar","o" = "kar" + b + "" = karb
  replaces   = [L + c + R[1:]           for L, R in splits if R for c in alphabets]
  # print(replaces)

  # "","karo" = "" + b + "karo" = bkaro 
  # "k","aro" = "k" + b + "aro" = kbaro
  # "ka","ro" = "ka" + b + "ro" = kabro
  # "kar","o" = "kar" + b + "o" = karbo
  # "karo","" = "karo" + b + "" = karob
  inserts    = [L + c + R               for L, R in splits for c in alphabets]
  # print(inserts)



  # error[word] = set(deletes + transposes + replaces + inserts)
  # print("\n\n")
  # word,deletes + transposes + replaces + inserts
  # error.add()
  return set(deletes + transposes + replaces + inserts)
# print(error.get('ka'))
# total_corpus = sum(uni.values())
# print(uni['hbana'] / total_corpus)
idst_table('karo')
```

```python
def error_model(word):
  error = idst_table(word)
  # for key in list(error.keys()):
  #   if error[key].issubset(word):
  print('With Modeled Table = ',word, ":", len(error))
  all = len(error)
  # all = prob(word)

  # Incorrect Cka  = ka : [cka, kaz, ]
  for key in list(miss_all.keys()):
    for find in miss_all[key]:
      if find == word:
        print('With missmatched.txt = ',word, ":", len(miss_all[key]))
        selected = len(miss_all[key])
        original_prob = prob(key)
        original_word = key

  print('Correct Word = ',original_word, ":", original_prob)
  probs = selected/all
  return probs

print(error_model('cka'))
```

## Candidate Words Generation

```python
def candidates(word): 
  # print((known([word])))
  # print(known(idst_table(word)))
  # print([word])
  # return (known([word]) or known(idst_table(word)) or [word])
  misspell = idst_table(word)
  # corpset = set(w for w in words if w in uni)
  finder = []
  words = [word]
  for w in uni:
    if w in words:
      # print(w)
      finder.append(w)
    if w in misspell:
      # print(w)
      finder.append(w)
  # print(set(finder))
  return (set(finder))
  # print((known([word])))
  # print(known(idst_table(word)))
  # print([word])
  # ka
  # return (known([word]) or known() or [word])
# candidates('karao')
```

## Selection Model Using argmax

```python
def argmax(word):
  possible_words = candidates(word)
  cand_prob = dict()
  for selected_candidate in possible_words:
    cand_prob[selected_candidate] = prob(selected_candidate)
  # print(cand_prob)
  max_key = max(cand_prob, key=cand_prob.get)
  # print("Max Key = ",max_key)
  # print("General Argmax = ",max(candidates(word), key=prob))
  # return max(candidates(word), key=prob)
  return max_key
```
